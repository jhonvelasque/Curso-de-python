```python

# COMO TRAER TODOS LOS ELEMENTOS DENTRO DE UNA CARPETA
# (y moverlos a una carpeta nueva)

# carpeta destino donde mover los archivos
dest = path.parent / 'archivos_movidos'
dest.mkdir(exist_ok=True)

for archivo in path.parent.glob('*.xlsx'):
    destino = dest / archivo.name
    archivo.rename(destino)
    print('Movido:', archivo, '->', destino)
```

```python
# EJEMPLO: copiar todos los archivos .xlsx a otra carpeta (sin borrar el original)

origen = path.parent
copia_destino = origen / 'archivos_copiados'
copia_destino.mkdir(exist_ok=True)

for archivo in origen.glob('*.xlsx'):
    destino = copia_destino / archivo.name
    shutil.copy2(archivo, destino)  # copy2 preserva metadatos (timestamps)
    print('Copiado:', archivo, '->', destino)
```

```python
# COMPARACIÓN: FORMA TRADICIONAL vs PATHLIB

# ===== FORMA TRADICIONAL (strings + os) =====
import os

ruta_tradicional = r'D:\Users\jvelasqu\Desktop\PROYECTOS\curso de python\Parte 8\Data\demo\tradicional.txt'

# Crear carpetas si no existen
os.makedirs(os.path.dirname(ruta_tradicional), exist_ok=True)

# Escribir en archivo
with open(ruta_tradicional, 'w') as f:
    f.write('Forma tradicional con strings y os module.\n')

print('Archivo creado forma tradicional ✓')

# ===== FORMA MODERNA (Pathlib) =====
from pathlib import Path

ruta_moderna = Path(r'D:\Users\jvelasqu\Desktop\PROYECTOS\curso de python\Parte 8\Data\demo\moderna.txt')

# Crear carpetas si no existen
ruta_moderna.parent.mkdir(parents=True, exist_ok=True)

# Escribir en archivo
with ruta_moderna.open('w') as f:
    f.write('Forma moderna con Pathlib.\n')

print('Archivo creado forma moderna ✓')

# ===== COMPARACIÓN DE OPERACIONES =====
print('\n--- COMPARACIÓN DE SINTAXIS ---')
print(f'Tradicional - existe: {os.path.exists(ruta_tradicional)}')
print(f'Pathlib - existe: {ruta_moderna.exists()}')

print(f'\nTradicional - nombre: {os.path.basename(ruta_tradicional)}')
print(f'Pathlib - nombre: {ruta_moderna.name}')

print(f'\nTradicional - carpeta padre: {os.path.dirname(ruta_tradicional)}')
print(f'Pathlib - carpeta padre: {ruta_moderna.parent}')
```
